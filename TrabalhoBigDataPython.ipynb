```python
import pandas as pd
```

```python
teste = pd.read_excel("TrabalhoVRaphael.xlsx")

display(teste)
```

```python
import pandas as pd
for column in teste.columns:
  if column not in ["Município", "Localização", "Dependência"]:
    teste[column] = pd.to_numeric(teste[column], errors='coerce')

display(teste)
```

```python
#TOTAL DE TODOS OS MUNICÍPIOS DO ENSINO FUNDAMENTAL

import pandas as pd

municipio_means = teste.groupby('Município').agg({'TotalAP': 'mean', 'TotalRep': 'mean', 'TotalAB': 'mean'})

municipio_means = municipio_means.round(2)

print(municipio_means)
```

```python
#TOTAL DE TODOS OS MUNICÍPIOS DO ENSINO MÉDIO

municipio_means_total = teste.groupby('Município').agg({'TotalEMAP': 'mean', 'TotalEMREP': 'mean', 'TotalEMAB': 'mean'})

municipio_means_total = municipio_means_total.round(2)

print(municipio_means_total)
```

```python
#TOTAL DE TODOS OS MUNICÍPIOS

media_aprovacao_total = teste['TotalAP'].mean()
media_reprovacao_total = teste['TotalRep'].mean()
media_abandono_total = teste['TotalAB'].mean()

print(f"Média de aprovação total: {media_aprovacao_total:.2f}")
print(f"Média de reprovação total: {media_reprovacao_total:.2f}")
print(f"Média de abandono total: {media_abandono_total:.2f}")
```

```python
#MOSTRA AS MAIORES MÉDIAS DE ABANDONO

abandono_por_municipio = teste.groupby('Município')['TotalAB'].mean()

abandono_por_municipio_ordenado = abandono_por_municipio.sort_values(ascending=False)

print("Municípios com maior média de abandono escolar:")
print(abandono_por_municipio_ordenado.head(10))
```
